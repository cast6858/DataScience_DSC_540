```python
#copying data frame to know difference between dup drop and mean conversion
met_data_mean_cal = met_data_drop_objDup.copy()
```

Filling in missing date for accession year by mean

```python
#making sure that there are no values that arent a number inside of the accession year. If there is a numeric error


#returns numeric values and remove non numeric values
met_data_Accession_Fill = met_data[met_data.AccessionYear.apply(lambda x: str(x).isnumeric())]


#if its a number getting the mean years 
mean_yearTo_Fill = round(met_data_Accession_Fill.AccessionYear.astype(int).mean())


# filling in all null values
met_data_mean_cal['AccessionYear'].fillna(str(mean_yearTo_Fill))


met_data_mean_cal.head()

                                      
```
